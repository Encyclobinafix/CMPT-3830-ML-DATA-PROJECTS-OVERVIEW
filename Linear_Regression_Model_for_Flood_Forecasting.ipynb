# **Dataset Description :**
This dataset contains information on various factors that may influence flood risk. The dataset includes several features representing environmental, social, infrastructure, and governance-related factors that could impact the likelihood and severity of flooding events. Additionally, it includes a target variable, "FloodProbability," which may indicate the likelihood of flooding occurring in different areas within the region.

----

# **Columns :**
- **MonsoonIntensity:** This column likely represents the intensity or strength of monsoon rains in a particular region. Monsoon intensity, values from  ( `0  to 16` ).
- **TopographyDrainage:** Topography refers to the surface features of a region, and drainage refers to the system of water flow, including rivers, streams, and drainage       channels, values from ( `0  to 18` ).
- **RiverManagement:** This column could represent the management practices implemented along rivers, such as flood control measures, dam operations, or river channel           modifications, values from ( `0  to 16` ).
- **Deforestation:** This column likely reflects the extent of deforestation in the region. Deforestation can affect the natural water cycle, soil stability, and increase the   risk of erosion and flooding, values from ( `0  to 17` ).
- **Urbanization:** Urbanization refers to the process of population concentration in urban areas and the expansion of urban land use, values from ( `0  to 17` ).
- **ClimateChange:** This column may represent factors related to climate change, such as changes in precipitation patterns, temperature variability, or extreme weather         events, values from ( `0  to 17` ).
- **DamsQuality:** This column could indicate the condition or quality of dams in the region. Dams play a significant role in water management, including flood control, water   supply, and hydropower generation, values from ( `0  to 16` ).
- **Siltation:** Siltation refers to the accumulation of sediment (silt, sand, and clay particles) in water bodies, such as rivers, lakes, and reservoirs, values from ( `0  to 16` ).
- **AgriculturalPractices:** This column may represent agricultural practices and land use patterns in the region. Agricultural activities can impact soil erosion, water       infiltration, and runoff, affecting flood risk, values from ( `0  to 16` ).
- **Encroachments:** Encroachments refer to unauthorized occupation or development on public or private land, such as illegal construction in floodplains or riverbanks,          values from ( `0  to 16` ).
- **IneffectiveDisasterPreparedness:** This column likely represents the level of effectiveness or readiness of disaster preparedness measures in the region, values from ( `0  to 16` ).
- **DrainageSystems:** This column may indicate the condition or effectiveness of drainage systems in the region. Drainage systems include both natural  and man-made  systems   designed to manage surface water runoff and reduce the risk of flooding, values from ( `0  to 17` ).
- **CoastalVulnerability:** This column could represent the vulnerability of coastal areas to flooding or other coastal hazards, such as storm surges, sea level rise, or        erosion, values from ( `0  to 17` ).
- **Landslides:** This column likely reflects the susceptibility or frequency of landslides occurring in the region , values from ( `0  to 16` ).
- **Watersheds:** This column may indicate the characteristics or condition of watersheds within the region. Watersheds are areas of land where all water drains to a common      point, such as a river or lake, values from ( `0  to 16` ).
- **DeterioratingInfrastructure:** This column could represent the condition or quality of infrastructure such as roads, bridges, and drainage systems. Deteriorating           infrastructure may be more prone to failure during extreme weather events, exacerbating flood impacts, values from ( `0  to 17` ).
- **PopulationScore:** This column may represent a measure of population density or population distribution within the region. High population density in flood-prone areas     can increase the vulnerability of communities to flood hazards, values from ( `0  to 18` )
- **WetlandLoss:** This column likely reflects the extent or rate of wetland loss within the region. Wetlands provide valuable ecosystem services, including flood regulation,   water purification, and habitat for biodiversity, values from ( `0  to 19` )
- **InadequatePlanning:** This column may indicate deficiencies or shortcomings in urban or land use planning practices that contribute to increased flood risk, values from ( `0  to 16` )
- **PoliticalFactors:** This column could represent political or governance-related factors that influence flood risk management and mitigation efforts, values from ( `0  to 16` )
- **FloodProbability:** This column likely represents the probability or likelihood of flooding occurring in the region, values from (`0.245 to 0.725` )

**4.1 Histogram Distribution of Flood Dataset**

# **Step 5 : Split data**
---

```python
X = df.drop(columns= ['FloodProbability']) #features
y = df['FloodProbability'] #target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.85,random_state=0)
```

# **Step 6 : Train the Model**
------

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

# **Step 7 : Predictions**
---

```python
y_pred_lr = lr.predict(X_test)
```

# **Step 8 : Accuracy of Model**
---

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
```

```python
mape = mean_absolute_percentage_error(y_test,y_pred_lr)
print("Error of Linear Regression Model = %.2f"%(mape*100),'%')
print("Accuracy of Linear Regression Model = %.2f"%((1 - mape)*100),'%')
```

```python
r2 = r2_score(y_test,y_pred_lr)
print("R2 score of Linear Regression = %.2f"%(r2))
```
